```julia
y0=0.15
wn=sqrt(2)
zeta=1/(2*sqrt(2));
t=LinRange(0,10,101)
```

```julia
# unforced.m 
# Compute Unforced Response to an Initial Condition

c=(y0 / sqrt(1-zeta^2))

y=c*exp.(-zeta*wn*t).*sin.(wn.*sqrt.(1-zeta^2).*t.+acos(zeta))

bu=c.*exp.(-zeta*wn*t);bl=-bu;
```

```julia
# import Pkg # Needs to be executed one time after installing Julia to get the plotting package. 
# Pkg.add("Plots")
# Pkg.add("LaTeXStrings")
# Pkg.add("PyPlot")
pyplot()
using Plots
using LaTeXStrings
plot(t,y, title =L"Response versus time", label="response")
plot!(t,bu, label = string("upper portion of decay envelope: " , L"\frac{y(0)}{\sqrt{1-\zeta^2}}e^{-\zeta\omega_{n}t}"))
plot!(t,bl, label = string("lower portion of decay envelope: " , L"-\frac{y(0)}{\sqrt{1-\zeta^2}}e^{-\zeta\omega_{n}t}"))
```
